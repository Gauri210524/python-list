### Q43. Convert list to tuple and set.

```python
# Write your answer here
prices_with_gst = [11, 29, 59, 65]

prices_tuple = tuple(prices_with_gst)
prices_set = set(prices_with_gst)

print("Tuple:", prices_tuple)
print("Set:", prices_set)
```
